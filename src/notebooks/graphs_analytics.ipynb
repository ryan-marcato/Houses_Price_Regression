# Graficos

## Histograma

```python
col_plot = ['SALE_PRC', 'age']
fig, axes= plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(22, 8)
fig.suptitle('Histogramas sobre preÃ§o e idade', fontsize=24, y=1.03)

coluna = 0

for i in col_plot:
    ax = sns.histplot(x=df[i], kde=True, ax=axes[coluna], color='g')
    ax.set_title(i, fontsize=22, y=1.02)
    
    coluna += 1
    
    plt.tight_layout()
    
```

## Graficos de barra

```python
ax = sns.barplot(df['structure_quality'].value_counts(), palette='pastel')
ax.figure.set_size_inches(14, 6)
ax.set_xlabel('Qualidade da estrutura', fontsize=18)
ax.set_ylabel('')

plt.show()
```
