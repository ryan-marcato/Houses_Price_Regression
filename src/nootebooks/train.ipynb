```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from feature_engine import pipeline

from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor
```

```python
import warnings

warnings.filterwarnings("ignore")
```

```python
df_train = pd.read_csv('..//data//processing//train.csv')

df_train
```

```python
df_test = pd.read_csv('..//data//processing//test.csv')

df_test
```

```python
df_train.drop(['Unnamed: 0'], axis=1, inplace=True)
df_test.drop(['Unnamed: 0'], axis=1, inplace=True)

df_train
```

```python
features = df_train.drop(columns='SALE_PRC').columns.to_list()
target = 'SALE_PRC'

X_train = df_train[features]
y_train = df_train[target]
X_test = df_test[features]
y_test = df_test[target]
```

# LINEAR REGRESSION

```python
linear_regression = LinearRegression()
```

## GridSearchCV

```python
params = {
    'fit_intercept': [True, False],
}

grid = GridSearchCV(linear_regression,
                    param_grid=params,
                    cv=5,
                    n_jobs=-1)
```

## Pipeline

```python
pipeline = pipeline.Pipeline([
    ('scaler', StandardScaler()),
    ('grid', grid)
])
```

## Treino

```python
pipeline.fit(X_train, y_train)
```

```python
predict_train = pipeline.predict(X_train)
predict_test = pipeline.predict(X_test)
```

```python
train_r2 = metrics.r2_score(y_train, predict_train)
test_r2 = metrics.r2_score(y_test, predict_test)
train_rmse = metrics.root_mean_squared_error(y_train, predict_train)
test_rmse = metrics.root_mean_squared_error(y_test, predict_test)
```

```python
print(f'Train R2: {round(train_r2 * 100, 2)}')
print(f'Test R2: {round(test_r2 * 100, 2)}')
print(f'Train RMSE: {round(train_rmse, 2)}')
print(f'Test RMSE: {round(test_rmse, 2)}')
```

# Naive bayes

```python
nb = GaussianNB()
```

## GridSearchCV

```python
params = {
    'var_smoothing': [1e-9, 1e-10, 1e-11]
}

grid = GridSearchCV(nb,
                    param_grid=params,
                    cv=5,
                    n_jobs=-1)
```

## Pipeline

```python
pipeline = pipeline.Pipeline([
    ('scaler', StandardScaler()),
    ('grid', grid)
])
```

```python
predict_train = pipeline.predict(X_train)
predict_test = pipeline.predict(X_test)
```

```python
train_r2 = metrics.r2_score(y_train, predict_train)
test_r2 = metrics.r2_score(y_test, predict_test)
train_rmse = metrics.root_mean_squared_error(y_train, predict_train)
test_rmse = metrics.root_mean_squared_error(y_test, predict_test)
```

```python
print(f'Train R2: {round(train_r2 * 100, 2)}')
print(f'Test R2: {round(test_r2 * 100, 2)}')
print(f'Train RMSE: {round(train_rmse, 2)}')
print(f'Test RMSE: {round(test_rmse, 2)}')
```
