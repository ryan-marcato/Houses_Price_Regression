```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("..//data//archive//miami-housing.csv")

df
```

```python
round(df.describe(), 2)
```

# Graficos

## Histograma

```python
col_plot = ['SALE_PRC', 'age']
fig, axes= plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(22, 8)
fig.suptitle('Histogramas sobre preÃ§o e idade', fontsize=24, y=1.03)

coluna = 0

for i in col_plot:
    ax = sns.histplot(x=df[i], kde=True, ax=axes[coluna], color='g')
    ax.set_title(i, fontsize=22, y=1.02)
    
    coluna += 1
    
    plt.tight_layout()
    
```

## Graficos de barra

```python
ax = sns.barplot(df['structure_quality'].value_counts(), palette='pastel')
ax.figure.set_size_inches(14, 6)
ax.set_xlabel('Qualidade da estrutura', fontsize=18)
ax.set_ylabel('')

plt.show()
```

```python
colors = sns.color_palette('Oranges', n_colors=12)

ax = sns.barplot(df['month_sold'].value_counts(), palette=colors)
ax.figure.set_size_inches(14, 6)
ax.set_title('Quantidade de vendas de cada mes', fontsize=24, y=1.03)
ax.set_xlabel('Mes', fontsize=18)
ax.set_ylabel('Quantidade', fontsize=18)

plt.show()
```

## Grafico de dispersÃ£o

```python
x_plots = df.drop(columns='SALE_PRC').columns.to_list()

ax = sns.pairplot(data=df, y_vars='SALE_PRC', x_vars=x_plots, kind='reg')
ax.figure.set_size_inches(30, 3)

ax.tight_layout()
plt.show()
```
